```python
from sklearn import *
import pandas as pd

data = pd.read_csv('23OCT_23NOV_10.193.123.18_final.csv.csv', header=0, encoding='UTF-16')
print(data.head())

data.describe
```

```python
numericdata = data._get_numeric_data()
categorical = data.drop(numericdata.columns, axis=1)
```

```python
data1 = data.apply(preprocessing.LabelEncoder().fit_transform)
y = data1['PacketDrop_Current_Value']
X = data1.drop(['PacketDrop_Current_Value','PacketDrop_Severity'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42, shuffle=True)
```

```python
from sklearn import linear_model
clf = linear_model.SGDClassifier(max_iter=1000)
```

```python
clf.fit(X_train, y_train)
```

```python
my_op = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(my_op,y_test)
```
